```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.models import load_model
import pickle
```

```python
## Loading the files :- 
# trained model
# scaler pickle file
#one hot encoder pickle file

model = load_model('model.h5')

with open('scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)

with open('label_encoder_gender.pkl', 'rb') as f:
    label_encoder = pickle.load(f)

with open('ohe_geography.pkl', 'rb') as f:
    ohe_geography = pickle.load(f)
    
```

```python
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
geo_endcoded = ohe_geography.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_endcoded, columns=ohe_geography.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

```python
## Encode the categorical features
input_df['Gender'] = label_encoder.transform(input_df['Gender'])
```

```python
input_df
```

```python
input_df = pd.concat([input_df.drop(['Geography'], axis=1), geo_encoded_df], axis=1)
```

```python
## Scaling
input_scaler = scaler.transform(input_df)
```

```python
input_scaler
```

```python
prediction = model.predict(input_scaler)
```

```python
prediction
```

```python
prediction_proba = prediction[0][0]
```

```python
prediction_proba
```

```python
if(prediction_proba > 0.5):
    print("The customer is likely to leave the bank.")
else:
    print("The customer is likely to stay with the bank.")
```
